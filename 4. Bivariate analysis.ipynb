```python
# Variance Inference Factor
from statsmodels.stats.outliers_influence import variance_inflation_factor

def vif(x):
    vif=pd.DataFrame()
    vif['variables']=x.columns
    vif['VIF'] = [variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return(vif)
```

```python
vif(dataset[[  'age',
               'blood_pressure',
               'specific_gravity',
               'albumin',
                'sugar',
                'red_blood_cells',
                'pus_cell',
                'pus_cell_clumps',
                'bacteria',
                'blood_glucose_random',
                'blood_urea',
                'serum_creatinine',
                'sodium',
                'potassium',
                'haemoglobin',
                'packed_cell_volume',
                'white_blood_cell_count',
                'red_blood_cell_count',
                'hypertension',
                'diabetes_mellitus',
                'coronary_artery_disease',
                'appetite',
                'peda_edema',
                'aanemia',
                'class'
               ]])
```

```python
# t-test is of two types namely,
               # 1. independent sample-unpaired test (different groups but same condition)
               # 2. depependent sample-paired test (same group but different condition)
import numpy as np
from scipy.stats import ttest_ind

data = {
    'blood_pressure': dataset['blood_pressure'],
    'hypertension': dataset['hypertension']
}

# Perform t-test for "bp"
t_stat_bp, p_value_bp = stats.ttest_ind(data['hypertension'], data['blood_pressure'])

# Print results
print("t-statistic:", t_stat_bp)
print("p-value:", p_value_bp)

if p_value_bp < 0.05:
    print("There is a significant difference between hypertension and blood_pressure.")
else:
    print("There is no significant difference between hypertension and blood_pressure.")
 
```
