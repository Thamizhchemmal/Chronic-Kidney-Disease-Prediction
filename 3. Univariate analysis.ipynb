```python
def cat_num(dataset):
    categorical_values=[]
    numerical_values=[]
    for i in dataset.columns:
        print(i)
        if(dataset[i].dtype=='O'):
            categorical_values.append(i) 
        else:
            numerical_values.append(i)
    return categorical_values,numerical_values
```

```python
# Calculation of univariate parameters

# 1.Mean: It is a summation of all datas. 
# purpose: For comparing the datas.

# 2.Median: # It is a middle value in a dataset when it is ordered from lowest to highest (or highest to lowest).
#             It is the value that separates the higher half of the data from the lower half. 
# purpose: To omit outliers

# 3.Mode: Represents the most frequently occurring data.
# purpose: To make a decision by taking the most common data.

# percentile: Describes the relative position of a particular value within a dataset.
#             It divides a dataset into 100 equal parts, with each part representing 1% of the data. 
#             It is commonly used to understand the distribution of data and identify how a specific value compares
#             to the rest of the values in the dataset.
# purpose: To know the location of data.

ckd_descriptive=pd.DataFrame(index=['Mean','Median','Mode','Q1:25%','Q2:50%',
                                    'Q3:75%','Q4:100%','IQR','1.5_Rule','Lesser_outliers','Greater_outliers',
                                    'Min','Max','kurtosis','skew','variance','std_deviation'],columns=numerical_values) 
for i in numerical_values:
    ckd_descriptive[i]['Mean']             = dataset[i].mean()
    ckd_descriptive[i]['Median']           = dataset[i].median()
    ckd_descriptive[i]['Mode']             = dataset[i].mode()[0]
    ckd_descriptive[i]['Q1:25%']           = dataset.describe()[i]['25%']
    ckd_descriptive[i]['Q2:50%']           = dataset.describe()[i]['50%']
    ckd_descriptive[i]['Q3:75%']           = dataset.describe()[i]['75%']
    ckd_descriptive[i]['Q4:100%']          = dataset.describe()[i]['max']
    ckd_descriptive[i]['IQR']              = ckd_descriptive[i]['Q3:75%']-ckd_descriptive[i]['Q1:25%']
    ckd_descriptive[i]['1.5_Rule']         = 1.5*ckd_descriptive[i]['IQR'] 
    ckd_descriptive[i]['Lesser_outliers']  = ckd_descriptive[i]['Q1:25%']-ckd_descriptive[i]['1.5_Rule'] 
    ckd_descriptive[i]['Greater_outliers'] = ckd_descriptive[i]['Q3:75%']+ckd_descriptive[i]['1.5_Rule'] 
    ckd_descriptive[i]['Min']              = dataset[i].min()
    ckd_descriptive[i]['Max']              = dataset[i].max()
    ckd_descriptive[i]['kurtosis']         = dataset[i].kurtosis()
    ckd_descriptive[i]['skew']             = dataset[i].skew()
    ckd_descriptive[i]['variance']         = dataset[i].var()
    ckd_descriptive[i]['std_deviation']    = dataset[i].std()
```

```python
# Identification
Lesser_outliers=[]
Greater_outliers=[]

for i in numerical_values:
    if(ckd_descriptive[i]['Min'] < ckd_descriptive[i]['Lesser_outliers']):
        Lesser_outliers.append(i)
    if(ckd_descriptive[i]['Min'] > ckd_descriptive[i]['Greater_outliers']):
        Greater_outliers.append(i)
    
```

# Frequency Table

```python
def Freq_table(i,dataset):
    Freq_table=pd.DataFrame(columns = ['Unique_values','Frequency','Relative_Frequency','Cumulative_Sum'])
    Freq_table['Unique_values']        = dataset[i].value_counts().index
    Freq_table['Frequency']            = dataset[i].value_counts().values
    Freq_table['Relative_Frequency']   = (Freq_table['Frequency']/2)
    Freq_table['Cumulative_Sum']       = dataset[i].cumsum()
    return Freq_table
```

```python
# Probability Function

def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    
    sample=dataset
    
    sample_mean = sample.mean()
    sample_std = sample.std()
    print('Mean =%3f,Standard Deviataion = %3f'%(sample_mean,sample_std))
    
    dist = norm(sample_mean,sample_std)
    
    values=[value for value in range(startrange,endrange)]
    probabilities = [dist.pdf(values) for value in values]
    prob = sum(probabilities)
    print('Area between range ({},{}),{}'.format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(dataset['blood_pressure'],50,80)
```

```python
# Standard Distribution Graph ( To convert to standard normal distribution)
def std_normal_dist_graph(dataset):
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    values=[i for i in dataset]
    z_score=[((j-mean)/std)for j in values]
    sns.distplot(z_score,kde=True)
    sum(z_score)/len(z_score)
    
```
