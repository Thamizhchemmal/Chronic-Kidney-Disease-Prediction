# SELECT K-MEANS ALGORITHM

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv('pre_processed_dataset.csv')
dataset
```

```python
dataset.isnull().sum()
```

```python
def selectkbest(x,y,n):
    test=SelectKBest(score_func=chi2,k=n)
    fit=test.fit(x,y)
    selectK_features=fit.transform(x)
    return selectK_features
```

```python
def split_scalar(x,y):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
    sc=StandardScaler()
    x_train=sc.fit_transform(x_train)
    x_test=sc.transform(x_test)
    return x_train,x_test,y_train,y_test
```

```python
def cm_prediction(classifier,x_test):
    y_pred=classifier.predict(x_test)
    from sklearn.metrics import confusion_matrix
    cm=confusion_matrix(y_test,y_pred)
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import classification_report
    accuracy=accuracy_score(y_test,y_pred)
    report=classification_report(y_test,y_pred)
    return classifier,accuracy,report,x_test,y_test,classifier
```

```python
def Logistic(x_train,y_train,x_test):
    from sklearn.linear_model import LogisticRegression
    classifier=LogisticRegression()
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def svm_linear(x_train,y_train,x_test):
    from sklearn.svm import SVC
    classifier=SVC(kernel='linear',random_state=0)
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def svmnl(x_train,y_train,x_test):
    from sklearn.svm import SVC
    classifier=SVC(kernel='rbf',random_state=0)
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def NaivesBayes(x_train,y_train,x_test):
    from sklearn.naive_bayes import GaussianNB
    classifier=GaussianNB()
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def KNN(x_train,y_train,x_test):
    from sklearn.neighbors import KNeighborsClassifier
    classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def DecisionTree(x_train,y_train,x_test):
    from sklearn.tree import DecisionTreeClassifier
    classifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def RandomForest(x_train,y_train,x_test):
    from sklearn.ensemble import RandomForestClassifier
    classifier=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def selectk_classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf):
    dataframe=pd.DataFrame(index=['chisquare'],columns=['Logistic',
                                                        'SVML',
                                                        'SVMNL',
                                                        'NaivesBayes',
                                                        'KNN',
                                                        'DecisionTree',
                                                        'RandomForest'])
    for number,index in enumerate(dataframe.index):
        dataframe['Logistic']    [index]  = acclog[number]
        dataframe['SVML']        [index]  = accsvml[number]
        dataframe['SVMNL']       [index]  = accsvmnl[number]
        dataframe['NaivesBayes'] [index]  = accnav[number]
        dataframe['KNN']         [index]  = accknn[number]
        dataframe['DecisionTree'][index]  = accdes[number]
        dataframe['RandomForest'][index]  = accrf[number]
    return dataframe
```

```python
df1=dataset
df1=pd.get_dummies(df1,drop_first=True)
df1
```

```python
x=df1.drop('class',axis=1)
y=df1['class']
```

```python
kbest = selectkbest(x,y,6)
acclog   = []
accsvml  = []
accsvmnl = []
accknn   = []
accnav   = []
accdes   = []
accrf    = []
```

```python
kbest
```

```python
x_train,x_test,y_train,y_test = split_scalar(kbest,y)
```

```python

classifier,accuracy,report,x_test,y_test,cm = Logistic(x_train,y_train,x_test)
acclog.append(accuracy)
classifier,accuracy,report,x_test,y_test,cm = svm_linear(x_train,y_train,x_test)
accsvml.append(accuracy)
classifier,accuracy,report,x_test,y_test,cm = svmnl(x_train,y_train,x_test)
accsvmnl.append(accuracy)
classifier,accuracy,report,x_test,y_test,cm = NaivesBayes(x_train,y_train,x_test)
accknn.append(accuracy)
classifier,accuracy,report,x_test,y_test,cm = KNN(x_train,y_train,x_test)
accnav.append(accuracy)
classifier,accuracy,report,x_test,y_test,cm = DecisionTree(x_train,y_train,x_test)
accdes.append(accuracy)
classifier,accuracy,report,x_test,y_test,cm = RandomForest(x_train,y_train,x_test)
accrf.append(accuracy)
```

```python
result = selectk_classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf)
result
```
